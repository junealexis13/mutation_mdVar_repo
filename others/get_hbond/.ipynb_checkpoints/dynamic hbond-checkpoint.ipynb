```python
import MDAnalysis as mda
from typing import Literal
from MDAnalysis.analysis.hydrogenbonds.hbond_analysis import HydrogenBondAnalysis as HBA
import glob, os , shutil
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import pandas as pd
import json
import seaborn as sns
```

```python
def get_hbond(traj: str, topol: str, accepter: Literal['whole','RBM','global'], da_distance = 3.2):
    # Load trajectory and topology using MDAnalysis
    u = mda.Universe(topol, traj, topology_format='TPR')

    # Define atom selections for donor and acceptor atoms (e.g., based on atom names or residue names)

    acceptor_selection = "resname UNK"   #Ligand resname
    donor_selection = "resid 107:174"   #RBM ResID
    donor_whole = "protein"

    # Calculate hydrogen bonds using MDAnalysis
    # Set cutoff distance to 3.5A (standard) 
    if accepter == 'whole':
        hbonds = HBA(u,
        donors_sel=donor_selection,
        acceptors_sel=acceptor_selection,
        d_a_cutoff=da_distance,
        between=[acceptor_selection,donor_whole],
        d_h_a_angle_cutoff=150,
        update_selections=True)

    elif accepter == 'RBM':
        hbonds = HBA(u, 
        donors_sel=donor_whole, 
        acceptors_sel=acceptor_selection, 
        d_h_a_angle_cutoff=150,
        between=[acceptor_selection,donor_selection],
        d_a_cutoff=da_distance,
        update_selections=True)
        
    elif accepter == 'global':
        hbonds = HBA(u, 
        d_h_a_angle_cutoff=165,
        between=["protein","resname UNK"],
        d_a_cutoff=da_distance,
        update_selections=True)
        
        protein_h =  hbonds.guess_hydrogens("protein")
        protein_acc = hbonds.guess_acceptors("protein")
        
        unk_h =  hbonds.guess_hydrogens("resname UNK")
        unk_acc = hbonds.guess_acceptors("resname UNK")
        
        
        hbonds.hydrogens_sel = f"({protein_h}) or ({unk_h})"
        hbonds.acceptors_sel = f"({protein_acc}) or ({unk_acc})"
        
        
    # Initiate HBOND - Analysis
    hbonds.run()
    
    return hbonds



def get_atoms(traj: str, topol: str, count_by_id = None, count_by_type = None):
    # Load trajectory and topology using MDAnalysis
    u = mda.Universe(topol, traj, topology_format='TPR')
    dynamic_int_hbonds = {}
    ncounter = 0
    for hbond_mdata in count_by_id:
#         if u.atoms[hbond_mdata[1]-1].element == "H"
        bond_interaction = []

        for i, info in enumerate(hbond_mdata):
            # the atom involved
            atom_ = u.atoms[info]
            if count_by_id is not None and count_by_type is None :
                if i <= 2:
                    bond_interaction.append(f"{atom_.resname}{atom_.resid-1+333}({atom_.element} - {info})")
                elif i == 3:
                    dynamic_int_hbonds[ncounter] = {
                        "bond_data": bond_interaction,
                        "count": int(info)
                    }

            elif count_by_id is None and count_by_type is not None:
                if i <= 1:
                    bond_interaction.append(info)
                elif i == 2:
                    dynamic_int_hbonds[ncounter] = {
                        "bond_data": bond_interaction,
                        "count": int(info)
                    }
        ncounter += 1

    return dynamic_int_hbonds

def matrix_heatmap(traj: str, topol: str):
    # Load trajectory and topology using MDAnalysis
    u = mda.Universe(topol, traj, topology_format='TPR')
    a = u.select_atoms("resid 107:174")
    return a
                     
def merge_substrings(lst: list):
    return '-'.join(lst)

def get_donor(lst: list):
    return f"{lst[0]}-{lst[2]}"

def get_res_donor(lst: list):
    return lst[0][:6]

def get_h(lst: list):
    return lst[1]

def reciprocal(x):
    return 1/x
```

```python
main_dir = os.path.join("/Users/junealexissantos/","Desktop","PRELIM_DATA")
all_tpr = glob.glob(os.path.join(main_dir,"*","*mdout.tpr"))
all_xtc = glob.glob(os.path.join(main_dir,"*","*_mdout_rotrans_for_MDAnalysis.xtc"))
```

```python
for i, (tpr, xtc) in tqdm(enumerate(zip(all_tpr, all_xtc))):
    code = os.path.basename(tpr).split("_")[0:3]
    fcode = "_".join(code)

    a = get_hbond(xtc, tpr, "global")
    b = get_atoms(xtc, tpr, count_by_id = a.count_by_ids())

    with open(os.path.join(main_dir,f"{fcode}_HB_Profile.json"),"w") as rd:
        json.dump(b, rd, indent=3)
        rd.close()
        break
        
```

```python
#batch read all json files and plot it
#all bond_data is in D-H-a
codes = ['alpha ','E484A ','N440K ','N501Y ','Q493R ','Q498R ','S477N ',"T478K ",'Y505H ','omicron ']
fig,ax = plt.subplots(figsize=(20,12), ncols=5, nrows=2, dpi=300)

r = 0
c = 0
for i,json_file,label in zip(range(len(all_json)),all_json,codes):
    with open(json_file,"r") as rd:
        js = json.load(rd)
        df = pd.DataFrame.from_dict(js, orient='index')
        df['bond_data'] = df['bond_data'].apply(get_donor)
        

        #programmatic plotting
        length = max(df['count'])
        numbers = len(df['bond_data'])
        
        if numbers < 20:
            lsize = 8
            adjust = 0
        else:
            lsize= 5
            adjust = 0.25
        
        ax[r,c].set_xlim(0,length+length*0.5)
        
        ax[r,c].tick_params(axis='y', which='major',labelsize=lsize)
        ax[r,c].barh(df.index,df['count'], color='orange', label=label+'hbond')
        
        for ii, (x,y) in enumerate(zip(df['bond_data'],df['count'])):
                ax[r,c].text(y+y*0.05,ii-adjust,x + f", {str(y)}",fontsize=3-numbers*0.005 )
                
        ax[r,c].set_ylim(-.5, numbers-.5)
        ax[r,c].legend(fontsize=6)
        c += 1
        if i == 4:
            r+=1
            c = 0
        
    rd.close()
        
plt.subplots_adjust(wspace=0.2, hspace=0.1)
plt.savefig(os.path.join(main_dir,'dynamic_hydrogen_bonding_bydonortype.jpg'),dpi=300)
```

```python
fig, ax = plt.subplots(figsize=(22, 22), ncols=1, nrows = 10)  # Set the figure size
for i, (tpr, xtc) in tqdm(enumerate(zip(all_tpr, all_xtc))):
    reses = matrix_heatmap(xtc, tpr)
    name = os.path.basename(tpr)
    title_name = name.split("_")[0:3]
    res_bucket = []
    for x in range(len(reses)):
        resname = f"{reses.atoms[x].resname}{reses.atoms[x].resid + 333 - 1}"
        if resname not in res_bucket:
            res_bucket.append(resname)
    if "omic" in title_name:
        res_bucket.append("UNK525")
    elif "omic" not in title_name:
        res_bucket.append("UNK527")     
        
    #constructing a template PD Dataframe
    df = pd.DataFrame(index=["UNK527 (DONOR)","UNK527 (ACCEPTOR)"], columns = res_bucket)
    df_converted = df.fillna(0)

    #filling the data in
    with open(os.path.join(main_dir, name.replace("mdout.tpr","HB_Profile.json")),"r") as rd:
        profile = json.load(rd)
        for hb in profile.values():
            donor_res = hb['bond_data'][0][:6]
            acceptor_res = hb['bond_data'][2][:6]
            count = int(hb['count'])
            
            #map the values
            #treat donor as index, acceptor as cols
            try:
                if "omic" not in title_name:
                    if donor_res == "UNK527" and acceptor_res != "UNK527":
                        df_converted.loc["UNK527 (DONOR)",acceptor_res] += count
                    elif donor_res != "UNK527" and acceptor_res == "UNK527":
                        df_converted.loc["UNK527 (ACCEPTOR)",donor_res] += count
                else:
                    if donor_res == "UNK525" and acceptor_res != "UNK525":
                        df_converted.loc["UNK527 (DONOR)",acceptor_res] += count
                    elif donor_res != "UNK525" and acceptor_res == "UNK525":
                        df_converted.loc["UNK527 (ACCEPTOR)",donor_res] += count
            except KeyError as e:
                print("ignoring keys as it was outside RBM",e)
                pass
            

        rd.close()
        
        
    sns.heatmap(df_converted, annot=True, cmap='RdYlGn_r', annot_kws={'size': 6}, ax=ax[i], norm=None, fmt='.0f', linewidth=0.5, cbar=False)
    ax[i].tick_params(which="major", axis="both", labelsize=7 )
    
    ax2 = ax[i].twinx()
    if name.split("_")[2] != "complex":
        ax2.set_ylabel(name.split("_")[2],rotation=270, verticalalignment='center')
    else:
        ax2.set_ylabel(name.split("_")[0],rotation=270, verticalalignment='center')
    if i == 9:
        ax[i].set_xlabel("Donor/Acceptor Residue ID")
    else:
        ax[i].set_xticks([])
plt.subplots_adjust(hspace=0)
plt.tight_layout()
plt.savefig(os.path.join(main_dir,f'unique_hbond_heatmap.jpg'),dpi=300)
plt.show()  # Display the plot  
```
